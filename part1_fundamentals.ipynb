#### Part I  .- Unsupervised vs Supervised Learning

Concepts:
- Input Variables.- features or predictors or independent variables.
- Output Variable.- know as target variable, dependent variable or response variable.
- Training set.- group of training instance or sample. 
- Cost function.- also known as error rate or value function.

##### Rules-Based vs. Machine Learning
######  The rules-based approach could fail because the target can change its behaviour to prevent the detection. Instead, the machine-learning approach is more cheaper because it learns form its own expereience and can also predict new kind of shapes of the target. 

##### Supervised vs Unsupervised
###### In supervised the AI agent has access to labels. In the unsupervised learning we don't have labels.
